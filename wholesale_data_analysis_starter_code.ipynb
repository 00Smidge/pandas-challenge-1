```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty

"""
  There may be times when it is best to create a new column and store
  the aggregated data of one of more columns
"""
df['total_units_price'] = df['unit_price'] * df['qty']
df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.

"""
  We need to know how much it is going to cost to ship our units and we need to store that
  shipping price. Using an if-else statement should make it possible to add the correct value to the new shipping_price column
"""
if df['unit_weight'].all() >= 50:
  df['shipping_price'] = df['unit_weight'] * 10
else:
  df['shipping_price'] = df['unit_weight'] * 7
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%

"""
  In cases where you are working with floats it may be necessary to reformat pandas.
  To do this you can use pd.options.display.float_format = '{:.2f}'.format
"""
df['total_price'] = df['unit_price'] + df['shipping_price'] * .0925
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

df['total_processing_cost'] = df.unit_cost + df.qty + df.shipping_price
```

```python
# Create a column for the profit of each line using line cost and line price
df['profit'] = df.unit_price - df.unit_cost
```

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
'''
  Find the total units for all clients, then return the top 5.
  First clean up and make a copy of the df, then groupby() can be used to
  group all the qty data points with by client_id. Then we can aggregate the
  qty(s) sum up their totals, sort the values and reset index
'''

top_five_totals = df.dropna(how='any').drop_duplicates().copy()
top_five_totals = top_five_totals.groupby('client_id').agg(
  unit_total = ('qty', 'sum')
  ).sort_values(by='unit_total', ascending=False).head(5).reset_index()
top_five_totals
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit.
'''
  For this problem we first need to make sure that we are comparing dataframes of the same length.
  The client_ids from the top_five_totals should be stored in a variable. With the top_five_ids the original
  dataframe can now be tested against to see if the client_ids from top_five_totals are in the original dataframe.
  If the condition check returns as true aggregate data from the original dataframe. The now aggregated data can be
  merged into the top_five_totals dataframe.
'''
# store top client_id(s)
top_five_ids = top_five_totals.client_id

'''
  Use .isin() to see if the client_id(s) from top_five_totals is in the original dataframe.
  If true then group data by id(s) and agg() datum.

  NOTE: agg() is used to aggregate datum related to a single column. In this case multiple columns
  need to be aggregated into one. One method to achieve this is to use lambda.
'''
# store aggregate data
agg_ttls = df[df['client_id'].isin(top_five_ids)].groupby('client_id').agg(
  total_shipping_price = ('shipping_price', 'sum'),
  total_revenue = ('unit_price', lambda x: (x * df.loc[x.index, 'qty']).sum()),
  total_profit = ('unit_price', lambda x: ((x - df.loc[x.index, 'unit_cost']) * df.loc[x.index, 'qty']).sum())
)

# merge agg_ttls
top_five_merged = top_five_totals.merge(agg_ttls, on='client_id').reindex()

# display top_five_totals
top_five_merged.style
```

```python

# Format the data and rename the columns to names suitable for presentation.
top_five_merged = top_five_merged.rename(columns={
  'client_id': 'Client ID', 
  'unit_total':	'Total Units Ordered', 
  'total_shipping_price':	'Total Spent on Shipping', 
  'total_revenue': 'Total Revenue Earned', 
  'total_profit': 'Total Profits Earned'})

# Define a function that converts a dollar amount to millions.
def amilli(x):
  return x / 1_000_000

# Rename the columns to reflect the change in the money format.
# Define the money columns.
top_five_merged['Total Spent on Shipping'] = top_five_merged['Total Spent on Shipping'].map('${:,.2f}'.format)
top_five_merged['Total Profit (millions)'] = amilli(top_five_merged['Total Profits Earned']).map('${:.2f}M'.format)
top_five_merged['Total Revenue (millions)'] = amilli(top_five_merged['Total Revenue Earned']).map('${:.2f}M'.format)
top_five_merged['Total Profits Earned'] = top_five_merged['Total Profits Earned'].map('${:,.2f}'.format)
top_five_merged['Total Revenue Earned'] = top_five_merged['Total Revenue Earned'].map('${:,.2f}'.format)

# Apply the currency_format_millions function to only the money columns. 
top_five_merged
```
